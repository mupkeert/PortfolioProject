**Preparing Data**

```python
data.columns
```

```python
X = data[['Income','MntWines','NumStorePurchases','MntGoldProds']]
```

```python
X
```

```python
y=data["NumWebPurchases"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
import sklearn
sklearn.metrics.SCORERS.keys()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
```

```python
k_value = {
    "n_neighbors" :[2,3,4,5]
    }
reg=KNeighborsRegressor()
```

```python
grid_s = GridSearchCV(reg, k_value, cv=3, scoring='neg_root_mean_squared_error')
grid_s.fit(X_train,y_train)
```

```python
print(grid_s.best_params_)
print("negative mse:"+ str(grid_s.best_score_))
```

```python
knn=KNeighborsRegressor(n_neighbors=grid_s.best_params_["n_neighbors"])
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
#print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
print("Mean Squared Error : ", mean_squared_error(y_test,y_pred,squared = True))
print("Root Mean Squared Error : ", mean_squared_error(y_test,y_pred,squared = False))
```
